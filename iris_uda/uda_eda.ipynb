## Exploratory Data Analysis (EDA)

#### Univariate Analysis (one variable)
###### Continuous Data


```python
# Categorical (spam mail)
# Continous (age)
```

```python
# Scatterplot for continuous UDA
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('iris.csv')
df
```

```python
x_axis = df.index
y_axis = df["petal.length"]
```

```python
sns.scatterplot(x=x_axis, y=y_axis, hue=df["variety"])
plt.title("Petal Length of Sampled Flowers")
plt.xlabel("Sample Number")
plt.ylabel("Petal Length")
```

```python
# Strip plot for continous UDA
variety = df["variety"]
```

```python
sns.stripplot(x=variety, y=y_axis, hue=variety)
plt.title("Petal Length Strip Plot of Sampled Flowers")
plt.xlabel("Sample Variety")
plt.ylabel("Petal Length")
```

```python
# Distribution plot for continous UDA
sns.distplot(y_axis)
plt.title("Petal Length Distribution\n Plot of Sampled Flowers")
plt.xlabel("Petal Length")
```

```python
# Histogram for continous UDA
plt.hist(y_axis)
plt.title("Petal Length Histogram\n of Sampled Flowers")
plt.xlabel("Frequency")
plt.ylabel("Petal Length")
```

```python
# Box plot for continous UDA
sns.boxplot(x=y_axis)
plt.title("Petal Lenght Box Plot\n of Sampled Flowers")
plt.xlabel("Petal Length")
```

###### Categorical Data

```python
# Count Plot for categorical UDA
sns.countplot(x=variety)
plt.title("Petal Variety Count Plot\n of Sampled Flowers")
plt.xlabel("Sample Variety")
plt.ylabel("Count")
```

```python
# Pie Chart for categorical UDA
labels = ["Setosa", "Versicolor", "Virginicia"]
plt.pie(variety.value_counts(), labels=labels)
plt.title("Petal Variety Pie Chart\n of Sampled Flowers")
```
