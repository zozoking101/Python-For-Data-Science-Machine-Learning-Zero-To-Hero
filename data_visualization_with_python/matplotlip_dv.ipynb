## Data Visualization Using Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Line plot
x_axis = [1, 2, 3]
y_axis = [i * 2 for i in x_axis]
plt.title("My First Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x_axis, y_axis, color = "red")
```

```python
# Line plots
x1_axis = [2, 4, 6, 8]
y1_axis = [1, 10, 100, 1000]
x2_axis = [8, 6, 4, 2]
y2_axis = [1, 10, 100, 1000]
plt.title("Two Line Plots")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x1_axis, y2_axis, color = "red")
plt.plot(x2_axis, y2_axis, color = "green")
plt.legend(['red', 'green'])
```

```python
# Histograms
values = [10, 15, 20, 10, 15]
plt.title("My First Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.hist(values, color="y", width=0.333)
```

```python
# Bar Charts
b_values = [10, 15, 20]
plt.title("Bar Plot")
plt.xlabel("Item")
plt.ylabel("Height")
plt.bar(b_values, [2, 1, 2], color="orange", width=1)
```

```python
# Pie Charts
labels = ["Blue", "Orange", "Green", "Red", "Purple", "Brown"]
explode = [0, 0, 0, 0.1, 0, 0.15]
colors = labels
p_values = [20, 40, 35, 25, 60, 30]
plt.title("Pie Chart")
plt.pie(p_values, labels=labels, explode=explode,colors=colors)
```

```python
x = [i for i in range(1, 11)]
y = [15, 12, 54, 49, 87, 75, 52, 14, 23, 1]
plt.title("Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.scatter(x, y, c=x, cmap="Accent")
```

```python
# Log plots
xs = [_ for _ in range(4)]
ys = [10 ** i for i in xs]
plt.title("Log Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.yscale('log')
plt.plot(xs, ys)
```

```python
# Polar plots
import numpy as np
r = 2
thetas = np.arange(0, (2 * np.pi), 0.01)
for radian in thetas:
    plt.polar(radian, r)
plt.title("Polar Plot")
```

#### Handling Dates

```python
import pandas as pd
import matplotlib.dates as mdates
dates = pd.Series(['2018-01-01', '2018-01-02', '2018-01-03', '2018-01-04', '2018-01-05', 
                  '2018-01-06', '2018-01-07', '2018-01-08', '2018-01-09', '2018-01-10'])
vs = [100, 101, 102, 103, 104, 105, 106, 107, 108, 109]
dates = pd.to_datetime(dates)
plt.title("Line Plot of Value against Time")
plt.xlabel("Dates")
plt.ylabel("Values")
# plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=90)
plt.plot(dates, vs)
```

#### Multiple Subplots in One Figure

```python
x1 = [2, 4, 6, 8]
y1 = [1, 10, 100, 1000]
x2 = [1, 3, 5, 7]
y2 = [100, 200, 300, 400]
```

```python
plt.subplot(1, 2, 1)
plt.title("Exponential\nLine Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x1, y1, c="r")

plt.subplot(1, 2, 2)
plt.title("Linear\nLine Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x2, y2, c="g")

plt.tight_layout()
```
